```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bound(D,d):
    avgs=[0]
    avg=0
    vec0=np.zeros((1,D.shape[0]))
    for i in range(1,d+1):
        r= np.random.choice([0, 1],size=(D.shape[0],1),p=[0.5,0.5])
        dr=D@r
        if(np.allclose(dr,vec0)):
            avg=avg+(1/i)*(1-avg)
        else:
            avg=avg+(1/i)*(0-avg)
        avgs.append(avg)
    return avgs,avg
```

```python
n=2
A=np.random.randint(0,10,size=(n,n))
B=np.random.randint(0,10,size=(n,n))
C=np.random.randint(0,10,size=(n,n))
D=(A@B)-C
print(f'A:\n {A}')
print(f'B:\n {B}')
print(f'C:\n {C}')
print(f'D:\n {D}')
d=1000
avgs,avg=bound(D,d)
print(avg)
plt.plot(range(d+1),avgs)
plt.plot(range(d+1),[0.5]*(d+1))
```
