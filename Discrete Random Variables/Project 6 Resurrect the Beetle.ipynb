# Resurrect The Beetle

```python
import numpy as np
from time import sleep
import matplotlib.pyplot as plt
```

```python
faces=12
legs=12
```

```python
def die():
    return int(np.random.uniform(1,faces+1))
```

```python
drawn=[0]*legs
turn=1
while sum(drawn)<6:
    roll=die()
    print(f'Turn: {turn}')
    print(f'\tRoll: {roll}')
    drawn[roll-1]=1
    print(f'\tLegs: {drawn}')
    turn+=1
    sleep(2)
print("Finish")
```

```python
def Beetle():
    drawn=[0]*legs
    turn=1
    while sum(drawn)<6:
        roll=die()
        drawn[roll-1]=1
        turn+=1
    return turn
```

```python
simulation=[Beetle() for i in range(10000)]
```

```python
avg=np.average(simulation)
plt.title("Turns to Finish")
plt.hist(simulation,density=True,bins=40)
print(f'Avg: {np.average(simulation)}')
print(f'STD: {np.std(simulation)}')
```

# Coupon Problem

The probability of obtaining a coupon $X_1 ~{} Geo(1)$

```python
numberCoupons=10
```

```python
Es=[1/(((numberCoupons+1)-d)/numberCoupons) for d in range(1,numberCoupons+1)]
plt.title("Average Boxes")
plt.xlabel("# of Different Coupons")
plt.ylabel("# of Boxes")
plt.bar(range(1,numberCoupons+1),Es)
```

```python
overall=[Es[0] if i==0 else Es[i]+Es[i-1] for i in range(numberCoupons)]
plt.title("Density Number of Boxes")
plt.xlabel("# of Different Coupons")
plt.ylabel("# of Boxes")
plt.bar(range(1,numberCoupons+1),overall)
```

One can assume that the number of boxes to open is the sum over the expected values given the memoryless property
